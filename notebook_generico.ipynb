```python
#Parte 1
from google.cloud import bigquery, storage
import os
import subprocess
import json

# Ruta al archivo de configuración
config_file_path = "C:\\Users\\TU_USUARIO\\Documents\\Apps\\config.json"  # Reemplaza con tu ruta

with open(config_file_path, 'r') as f:
    config = json.load(f)
    
os.environ['HTTPS_PROXY'] = config['network']['https_proxy']    


# Establece la variable de entorno para las credenciales de Google Cloud
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = config['google']['credentials']
#Ejecuta sp
from google.cloud import bigquery
'''
def ejecutar_procedimiento(project_id, dataset_id, stored_procedure_name):
    # Configura el cliente de BigQuery
    client = bigquery.Client(project=project_id)

    # Ejecuta el procedimiento almacenado mediante una consulta SQL
    sql_query = f"CALL {dataset_id}.{stored_procedure_name}()"
    query_job = client.query(sql_query)

    # Espera a que la consulta termine
    query_job.result()

    print(f"Procedimiento almacenado {stored_procedure_name} ejecutado correctamente.")
 

def main():
    # Configura tu información
    project_id = 'TU_PROJECT_ID'          # Reemplaza con tu ID de proyecto
    dataset_id = 'TU_DATASET_ID'          # Reemplaza con tu ID de conjunto de datos
    stored_procedure_name = 'NOMBRE_PROCEDIMIENTO'  # Reemplaza con el nombre de tu procedimiento almacenado

    # Llama a la función para ejecutar el procedimiento almacenado sin parámetros
    ejecutar_procedimiento(project_id, dataset_id, stored_procedure_name)

if __name__ == "__main__":
    main()

 '''  
def exportar_tabla_a_gcs(project_id, dataset_id, table_id, bucket_name, blob_name):
    # Configura el cliente de BigQuery y Storage
    client_bq = bigquery.Client(project=project_id)
    client_storage = storage.Client(project=project_id)

    # Referencia a la tabla en BigQuery
    table_ref = client_bq.dataset(dataset_id).table(table_id)

    # Configuración para la exportación
    extract_job_config = bigquery.ExtractJobConfig()
    extract_job_config.destination_format = bigquery.DestinationFormat.CSV
    extract_job_config.field_delimiter = ","  # Puedes ajustar el delimitador según tus necesidades

    # Ruta de destino en GCS
    destination_uri = f'gs://{bucket_name}/{blob_name}'

    # Ejecuta la exportación
    extract_job = client_bq.extract_table(
        table_ref,
        destination_uri,
        job_config=extract_job_config
    )

    # Espera a que la exportación termine
    extract_job.result()

    print(f"Exportación exitosa a {destination_uri}")

def main():
    # Configura tu información
    project_id = 'TU_PROJECT_ID'       # Reemplaza con tu ID de proyecto
    dataset_id = 'TU_DATASET_ID'       # Reemplaza con tu ID de conjunto de datos
    table_id = 'NOMBRE_TABLA'          # Reemplaza con tu ID de tabla
    bucket_name = 'NOMBRE_BUCKET'      # Reemplaza con el nombre de tu bucket
    blob_name = 'NOMBRE_ARCHIVO.csv'   # Reemplaza con el nombre del archivo de salida

    # Llama a la función para exportar la tabla a GCS
    exportar_tabla_a_gcs(project_id, dataset_id, table_id, bucket_name, blob_name)

if __name__ == "__main__":
    main()

#Parte 2

def descargar_archivo_desde_bucket():
    # Ruta del archivo en el bucket de Google Cloud Storage
    ruta_bucket = 'gs://NOMBRE_BUCKET/NOMBRE_ARCHIVO.csv'  # Reemplaza con tu ruta en GCS
    
    # Ruta local donde quieres descargar el archivo
    ruta_local = 'C:\\Users\\TU_USUARIO\\Documents\\EXPORT\\NOMBRE_ARCHIVO.CSV'  # Reemplaza con tu ruta local
    
    # Comando gsutil para copiar el archivo desde el bucket a la ubicación local
    comando_gsutil = f'gsutil cp {ruta_bucket} {ruta_local}'
    
    try:
        # Ejecuta el comando gsutil usando subprocess
        proceso = subprocess.Popen(comando_gsutil, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        stdout, stderr = proceso.communicate()
        
        # Verifica si la operación fue exitosa o si hubo algún error
        if proceso.returncode == 0:
            print("Descarga exitosa.")
        else:
            print("Error al descargar el archivo.")
            print("Salida de error:")
            print(stderr.decode())
    except Exception as e:
        print(f'Error: {str(e)}')

if __name__ == "__main__":
    descargar_archivo_desde_bucket()
```
